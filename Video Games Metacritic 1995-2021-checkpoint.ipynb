## Review meta_score

```python
bins = [0, 40, 60, 80, np.inf]
names = ['Bad', 'Average', 'Quite Good', 'Good']

all_games['review'] = pd.cut(all_games['meta_score'], bins, labels=names)
```

```python
all_games
```

```python
all_games.value_counts('review')
```

```python
plt.style.use("fivethirtyeight")

all_games.value_counts('review').plot(kind = 'pie', wedgeprops = {'edgecolor':'black'}, autopct = '%1.1f%%')
plt.ylabel("")
plt.title('Review')
```

## Year

```python
all_games['year'] = all_games['release_date'].dt.year
```

```python
all_games
```

```python
games_per_year = all_games['year'].value_counts().sort_index()
games_per_year = games_per_year.iloc[5:]
games_per_year = pd.DataFrame(games_per_year)
games_per_year
```

```python
games_after_2000 = all_games[~all_games['year'].isin([1995, 1996, 1997, 1998, 1999])]
games_after_2000 
```

```python
review_games = games_after_2000.groupby(['year', 'review'])['review'].count()
review_games 
```

```python
review_games.loc[:, 'Bad']
```

```python
average_score_per_year = all_games.groupby('year')['meta_score'].mean()
average_score_per_year = average_score_per_year.iloc[5:]
average_score_per_year
```

```python
games_per_year['Bad'] = pd.Series(review_games.loc[:, 'Bad'])
games_per_year['Average'] = pd.Series(review_games.loc[:, 'Average'])
games_per_year['Quite Good'] = pd.Series(review_games.loc[:, 'Quite Good'])
games_per_year['Good'] = pd.Series(review_games.loc[:, 'Good'])
games_per_year
```

```python
games_per_year.reset_index(inplace = True)
games_per_year
```

```python
games_per_year.rename(columns = {'Index' :'year', 'year':'number_of_games'}, inplace= True)
games_per_year
```

```python
fig, ax = plt.subplots(figsize = (20, 10))

ax.bar(x = games_per_year.index, height = games_per_year['Bad'],color = 'yellow', label = 'Bad')
ax.bar(x = games_per_year.index, height = games_per_year['Average'], bottom = games_per_year['Bad'],  color = 'red', label = 'Average')
ax.bar(x = games_per_year.index, height = games_per_year['Quite Good'], bottom = games_per_year['Bad'] + games_per_year['Average'], color = 'orange', label = 'Quite Good')
ax.bar(x = games_per_year.index, height =  games_per_year['Good'], bottom = games_per_year['Bad'] + games_per_year['Average'] + games_per_year['Quite Good'],
                                   color = 'blue', label = 'Good')

ax2 =  ax.twinx()
ax2.plot(games_per_year.index, average_score_per_year, label = 'average_score', color = 'black')

plt.grid()
plt.legend()
plt.xlabel('Year')
plt.title('Number of Games and Average Score per year')
```
